```python
folder = "dati-andamento-nazionale/"
filename = "dpc-covid19-ita-andamento-nazionale.csv"

dataset = pandas.read_csv(folder+filename, header=0)
dataset['data'] = dataset['data'].apply(datetime.datetime.fromisoformat)
dataset = dataset.set_index('data')
dataset.index = dataset.index.normalize()

dataset["incremento_casi_testati"] = dataset["casi_testati"].diff()

index_to_remove = [datetime.datetime.fromisoformat(i) for i in ["2020-12-05","2020-12-06","2020-12-07","2020-12-17","2020-12-18","2020-12-19"]]
dataset = dataset.drop(index=index_to_remove)
```

```python
dataset["incremento_casi_testati"]
```

```python
plt.figure(figsize=(15,6))
dataset["casi_testati"].plot()
dataset["totale_casi"].plot()
dataset["tamponi"].plot()
#plt.yscale("log")
```

```python
dataset["to_plot"] = dataset["nuovi_positivi"] / dataset["incremento_casi_testati"]
fig, ax = plt.subplots(1,1, figsize=(15,6))
seaborn.histplot(dataset, x="to_plot", bins=30, stat="density", kde=True)
xs = np.linspace(0, 0.4, 1001)
ax.plot(xs, beta.pdf(xs, 1.6, 65)/2 + beta.pdf(xs, 3, 10)/1.5, color="plum")

print(f"Our best estimator {xs[np.argmax(beta.pdf(xs, 1.6, 65)/2 + beta.pdf(xs, 3, 10)/1.5)]}")
```
