```python
## si prepara il dataset
output_dir = "output-grafici/"

regioni = pandas.read_csv("dati-regioni/dpc-covid19-ita-regioni.csv", header=0)
regioni['data'] = regioni['data'].apply(datetime.datetime.fromisoformat)
regioni = regioni.set_index('data')
regioni.index = regioni.index.normalize()
regioni.drop(['stato', 'codice_regione', 'lat', 'long', 'tamponi', 'note'], axis=1, inplace=True)

italia = pandas.read_csv("dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv", header=0)
italia['data'] = italia['data'].apply(datetime.datetime.fromisoformat)
italia = italia.set_index('data')
italia.index = italia.index.normalize()
italia.drop(['stato', 'tamponi', 'note'], axis=1, inplace=True)
italia = italia.assign( denominazione_regione = numpy.full(italia.shape[0], "Italia") )

dataset = pandas.concat([regioni,italia])

dataset.reset_index(inplace=True)
dataset = dataset.set_index(['denominazione_regione', 'data'])
dataset.sort_index(inplace=True)

areas = dataset.index.unique(level=0)

dataset = dataset.assign( incremento_casi_testati = numpy.zeros(dataset.shape[0]) )
for area in areas:
    dataset.loc[area, 'incremento_casi_testati'] = numpy.ediff1d(dataset.loc[area, 'casi_testati'], to_begin=numpy.nan)
```

```python
## si impostano i valori di interesse e si restringe il dataset
start_date = datetime.datetime.fromisoformat("2020-01-01") ### To be modified according to the wanted beginning of the period
finish_date = datetime.datetime.now()
wanted_areas = ['Italia','Lombardia','Veneto'] ### To be modified according to the wanted areas to be plotted

######## TO BE REMOVED
italia = italia.assign( incremento_casi_testati = numpy.ediff1d(italia['casi_testati'], to_begin=numpy.nan) )
regioni_plot = regioni[start_date:finish_date]
italia_plot = italia[start_date:finish_date]
########

restricted_dataset = dataset.loc[ (wanted_areas, slice(start_date,finish_date)), : ]
```

```python
## nuovi positivi / incremento testati
fig, axs = plt.subplots(1)
f = 100 * italia_plot['nuovi_positivi'] / italia_plot['incremento_casi_testati']
axs.plot(f)
axs.yaxis.set_major_formatter(matplotlib.ticker.PercentFormatter())
axs.set_title("Percentuale tamponi positivi")
fig.set_figwidth(12)
fig.tight_layout()
```

```python
## positivi, guariti, decessi / totali
fig, axs = plt.subplots(1,2)
f = 100 * italia_plot['totale_positivi'] / italia_plot['totale_casi']
axs[0].plot(f, label="Positivi")
f = 100 * italia_plot['dimessi_guariti'] / italia_plot['totale_casi']
axs[0].plot(f, label="Guariti")
f = 100 * italia_plot['deceduti'] / italia_plot['totale_casi']
axs[0].plot(f, label="Deceduti")
f = 100 * italia_plot['totale_casi'] / italia_plot['totale_casi']
axs[0].plot(f, label="Totale")
axs[0].yaxis.set_major_formatter(matplotlib.ticker.PercentFormatter())
axs[0].set_title("Ripartizione percentuale dei casi positivi")
axs[0].legend()

f = italia_plot['totale_positivi']
axs[1].plot(f, label="Positivi")
f = italia_plot['dimessi_guariti']
axs[1].plot(f, label="Guariti")
f = italia_plot['deceduti']
axs[1].plot(f, label="Deceduti")
f = italia_plot['totale_casi']
axs[1].plot(f, label="Totale")
axs[1].set_title("Casi positivi in valore assoluto")
axs[1].legend()

fig.set_figwidth(12)
fig.tight_layout()
```
